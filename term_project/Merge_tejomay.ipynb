```python
# 1. Load data
import pandas as pd
from os import listdir, getcwd
```

```python

gdelt = pd.read_csv('gdelt_dataframe.csv')

gdelt.head()
```

```python
# (b) Read in asylum seekers
seekers = pd.read_csv(getcwd() + '/Asylum_seekers_per_country_per_year_w_prev_year.csv')
seekers = seekers.drop('Unnamed: 0', axis = 1)
seekers.head()
```

```python
# (c) Read in populations
pop = pd.read_csv(getcwd() + '/populations.csv', skiprows = [i for i in range(0, 3)])
pop.head()
```

```python
# (d) Read in ISO codes
iso = pd.read_csv(getcwd() + '/ISO_codes.csv')
iso.head()
```

```python
# 2. Merge
# (a) seekers with iso
seekers_iso = seekers.merge(iso, left_on = 'Origin', right_on = 'English short name', validate = 'many_to_one', indicator = True)

print("Check seekers-iso merge:")
print(seekers_iso._merge.value_counts())

seekers_iso = seekers_iso.drop('_merge', axis = 1)

print(seekers_iso.head())
```

```python
# (b) merge seekers_iso with pop
seekers_iso_pop = seekers_iso.merge(pop, left_on = 'Alpha-3 code', right_on = 'Country Code', validate = 'many_to_one', indicator = True)

print("Check seekers-iso-pop merge:")
print(seekers_iso_pop._merge.value_counts())

seekers_iso_pop = seekers_iso_pop.drop('_merge', axis = 1)
print(seekers_iso_pop.head())
```

```python
# (b) merge seekers_iso_pop with gdelt
seekers_iso_pop.columns
seekers_iso_pop_gdelt = seekers_iso.merge(gdelt, left_on = ['Alpha-2 code', 'Year'], right_on = ['Actor2Geo_CountryCode', 'Year'], validate = 'many_to_many', indicator = True)

print("Check seekers-iso-pop-gdelt merge:")
print(seekers_iso_pop_gdelt._merge.value_counts())
print(seekers_iso_pop_gdelt.head())

seekers_iso_pop_gdelt = seekers_iso_pop_gdelt.drop('_merge', axis = 1)
```

```python
# 3. Export
seekers_iso_pop_gdelt.to_csv("merged_seekers_iso_pop_gdelt.csv")
```
